```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
plt.ion()

import sys
sys.path.append('/Users/jkinney/github/')
import logomaker
```

```python
# Load data
mat_df = pd.read_excel('PVIVIT_matrix_ddG.xlsx', index_col=0).fillna(0)
mat_df
```

```python
# Create figure and axes objects
fig, ax = plt.subplots(figsize=[2.5,2.5])

# Draw logo
logo = logomaker.Logo(-mat_df, 
                      colors='gray',
                      center=True, 
                      ax=ax, 
                      flip_below=False, 
                      font_family='Arial Rounded MT Bold')

# Highlight glyphs in sequence
# WARNING! NEED TO GET UPDATED MEASUREMETNS FROM POLLY. 
# AND HIGHLIGHTED SEQUENCE SHOULD BE '..PVIVIT..' I THINK
logo.style_glyphs_in_sequence('.FNYSQD.DZ', color='salmon')

# Turn all spines off expect the bottom spine
logo.style_spines(visible=False)
logo.style_spines(spines=['bottom'], visible=True)

# Use left spine to create a scale bar
logo.style_spines(spines=['left'], visible=True, linewidth=3, bounds=[3,4])

# Fade glyphs below x-axis
logo.style_glyphs_below(fade=.3)

# Do additional plot styling
L = len(mat_df)
ax.axhline(0,color='k',linewidth=.5, zorder=-1)
ax.set_xticks(range(L))
xticklabels = list(range(L))
xticklabels[0] = -1
ax.set_xticklabels(xticklabels)
ax.xaxis.set_tick_params(width=0)
ax.set_xlim([-1,L])
ax.set_yticks([])
ax.set_ylabel('$\Delta \Delta G$ (kcal/mol)')

# Tight layout
plt.tight_layout()

# Save as pdf
out_file = 'Example_Nguyen.pdf'
fig.savefig(out_file)
print('Done! Output written to %s.'%out_file)
```
