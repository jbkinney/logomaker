```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment(fasta_file='data/ww_domains.txt')
```

```python
parameters_txt = """
logo_type : 'probability'
below_flip : True
width : .9
axes_style : 'rails'
use_transparency : False
stack_order : 'small_on_top'
use_transparency : True
"""
with open('parameters.txt','w') as f:
    f.write(parameters_txt)
    
plt, ax = plt.subplots(1, figsize=[10,2])
logo = logomaker.make_styled_logo(style_file='parameters.txt', matrix=counts_mat)
logo.draw(ax)
```
